```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.log(np.arange(0.1, 100, 0.1))
y = 100/ (np.pow(np.e, -x+1)+1)
z = 40/ (np.pow(np.e, -x+1)+1)

plt.plot(x, y, label = 'Full Agonist')
plt.plot(x, z, label = 'Partial Agonist')
plt.axvline(x=1, color = 'gray', linestyle = '--', linewidth = 1)
plt.text(-1.5, 10, 'Same Potency (EC50)')
plt.axhline(40, color = 'gray', linestyle = '--', linewidth = 1)
plt.text(3, 105, 'Efficacy (Full Agonist)')
plt.axhline(100, color = 'gray', linestyle = '--', linewidth = 1)
plt.text(2.5, 45, 'Efficacy (Partial Agonist)')
plt.xticks(np.arange(-2, 5), [0.0001, 0.001, 0.01, 0.1, 1, 10, 100])
plt.legend()
plt.savefig('Full vs Partial Agonist.png', dpi = 600)
```

```python
x = np.log(np.arange(0.1, 100, 0.1))
y = 100/ (np.pow(np.e, -x+1)+1)
xz = np.log(np.arange(0.1, 1000, 0.1))
z = 100/ (np.pow(np.e, -xz+2.5)+1)

plt.plot(x, y, label = 'Agonist')
plt.plot(xz, z, label = 'Agonist \n+ Competitive Antagonist')
plt.axvline(x=1, color = 'gray', linestyle = '--', linewidth = 1)
plt.axvline(x=2.5, color = 'gray', linestyle = '--', linewidth = 1)
plt.text(1, 20, 'Shift in Potency')
plt.axhline(50, color = 'gray', linestyle = '--', linewidth = 1)
plt.text(3, 50, r'50% Efficacy')
plt.axhline(100, color = 'gray', linestyle = '--', linewidth = 1)
plt.text(3, 105, 'The Same Efficacy')
plt.xticks(np.arange(-2, 5), [0.0001, 0.001, 0.01, 0.1, 1, 10, 100])
plt.legend()
plt.savefig('Competitive Antagonist.png', dpi = 600)
```
