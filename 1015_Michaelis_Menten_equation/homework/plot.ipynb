```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams
import seaborn as sns
rcParams['font.family'] = 'serif'
rcParams['font.serif'] = ['Times New Roman']
```

```python

atp_conc = ['0', '0.03', '0.1', '0.3', '1', '3']
v_true = np.array([13.92199, 14.04078, 15.31843,	13.89617, 12.35871,	9.25853])
k_true = np.array([82.54831, 88.04961, 130.30483, 129.53799, 215.3569, 462.7184])
v_linear = np.array([12.2594091, 12.71294177, 11.79662616, 12.79754287, 11.31990038,6.618133686])
k_linear = np.array([53.29753586, 63.99389779, 60.83095435, 105.359355, 181.9314014,277.9775645])

fig, ax = plt.subplots(1, 2, figsize = (12, 5))
x = np.arange(len(atp_conc))
width = 0.35
alpha = 0.7
v_color = sns.color_palette('viridis', 12)
k_color = sns.color_palette('plasma', 12)

rects1 = ax[0].bar(
    x-width/2, v_true, 
    width, 
    alpha = alpha, facecolor = v_color[0::2])
rects2 = ax[0].bar(
    x+width/2, v_linear, 
    width, 
    alpha = alpha, facecolor = v_color[1::2])

line_mm_v = ax[0].plot(
    x-width/2, v_true, color = 'black',
    marker = 'o', label = 'Michaelis-Menten'
)
line_lb_v = ax[0].plot(
    x+width/2, v_linear, color = 'darkgrey',
    marker = 'x', label = 'Lineweaver-Burk'
)
ax[0].set_title('V$_{max}$')


rects3 = ax[1].bar(
    x-width/2, k_true, 
    width, 
    alpha = alpha, facecolor = k_color[0::2])
rects4 = ax[1].bar(
    x+width/2, k_linear, 
    width, 
    alpha = alpha, facecolor = k_color[1::2])

line_mm_k = ax[1].plot(
    x-width/2, k_true, color = 'black',
    marker = 'o', label = 'Michaelis-Menten'
)
line_lb_k = ax[1].plot(
    x+width/2, k_linear, color = 'darkgrey',
    marker = 'x', label = 'Lineweaver-Burk'
)
ax[1].set_title('K$_{m}$')

for i in range(2):
    ax[i].set_xticks(x)
    ax[i].set_xticklabels(atp_conc)
    ax[i].set_ylabel('Value')
    ax[i].set_xlabel('ATP Concentration (mmol/L)')
    ax[i].legend()

plt.savefig('MM_vs_LB.png', dpi = 600)
```

$${\frac {K_m}{V_{max}}}^*, {\frac 1 {V_{max}}}^* = \underset{\frac {K_m}{V_{max}}, \frac 1 {V_{max}}}{\operatorname{argmax}}\quad L (\frac {K_m}{V_{max}}, \frac 1 {V_{max}})$$
